```python
import pandas as pd

dataset = pd.read_csv('Melbourne_housing.csv')
dataset.head()
```

```python
dataset.nunique()
```

```python
dataset.shape
```

```python
#discard columns

cols_to_use = ['Suburb', 'Address', 'Rooms', 'Type', 'Price', 'Method', 'SellerG', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car', 'Landsize', 'BuildingArea', 'YearBuilt', 'CouncilArea', 'Lattitude', 'Longtitude']

dataset =  dataset[cols_to_use]
dataset.head()               
               
```

```python
dataset.shape
```

```python
dataset.isna().sum()
```

```python
cols_to_fill_zero = ['Price', 'Distance', 'Bedroom2', 'Bathroom']
dataset[cols_to_fill_zero] = dataset[cols_to_fill_zero].fillna(0)
dataset.isna().sum()
```

```python
dataset['Landsize'] = dataset['Landsize'].fillna(dataset.Landsize.mean())
dataset['BuildingArea'] = dataset['BuildingArea'].fillna(dataset.BuildingArea.mean())
```

```python
dataset.isna().sum()
```

```python
dataset.dropna(inplace = True)
dataset.isna().sum()
```

```python
dataset = pd.get_dummies(dataset,drop_first = True)
dataset.head()
```

```python
X = dataset.drop('Price', axis = 1)
y = dataset['Price']
```

```python
from sklearn.model_selection import train_test_split
train_X, test_X, train_y, test_y = train_test_split(X, y, test_size = 0.3, random_state = 2)
```

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression().fit(train_X, train_y)

                            
```

```python
reg.score(test_X, test_y)
```

```python
reg.score(train_X, train_y)
```

```python
from sklearn import linear_model

lasso_reg = linear_model.Lasso(alpha=50, max_iter=100, tol=0.1)

lasso_reg.fit(train_X, train_y)
```

```python
lasso_reg.score(test_X, test_y)
```

```python
lasso_reg.score(train_X, train_y)
```
